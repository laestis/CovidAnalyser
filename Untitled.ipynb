```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib import request
```

```python
request.urlretrieve ("https://raw.github.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv", "vaccinations.csv")
vaccinations=pd.read_csv("vaccinations.csv")
vaccinations.head()
```

```python
vaccinations[vaccinations.location=='France'].head()
```

```python
vaccinations.fillna(value={"total_vaccinations":0},inplace=True)
vaccinations.fillna(value={"people_vaccinated":0},inplace=True)
vaccinations.fillna(value={"people_fully_vaccinated":0},inplace=True)
vaccinations.fillna(value={"daily_vaccinations_raw":0},inplace=True)
vaccinations.fillna(value={"daily_vaccinations":0},inplace=True)
vaccinations.fillna(value={"total_vaccinations_per_hundred":0},inplace=True)
vaccinations.fillna(value={"people_vaccinated_per_hundred":0},inplace=True)
vaccinations.fillna(value={"people_fully_vaccinated_per_hundred":0},inplace=True)
vaccinations.fillna(value={"daily_vaccinations_per_million":0},inplace=True)
```

```python
vaccinations[vaccinations.location=='France'].head()
```

plt.bar(vaccinations[vaccinations.location=='France'].date,vaccinations[vaccinations.location=='France'].daily_vaccinations)
